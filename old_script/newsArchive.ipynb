```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.youtube.com/"

url = "https://www.youtube.com/"
finance_all = []
yahoo_r = requests.get(url)
yahoo_soup = BeautifulSoup(yahoo_r.text, 'html.parser')
finance = yahoo_soup.findAll('h3')
finance_all.append(finance)
len(finance_all)
print(yahoo_soup)
```

```python
len(finance_all)
links = []
base_url = "https://tw.news.yahoo.com"
for finance in finance_all:
    for info in finance:
        link = ""
        try:
            link = info.findAll('a', href=True)[0]
            title = link.get_text()
            print(title)
            # filter out 一覽表
            if title.find("一覽表") < 0 and link.get('href') != '#':
                full_url = base_url + link.get("href") 
                links.append( (title, full_url) )
                print(full_url)
                print('+++')
        except:
            print("exception")
            link = None
```

```python
mypath = './'
from os import listdir
from os.path import isfile, join
all_articles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
all_articles[:10]
```

```python
target_file = 'finance.txt'
count = 0
with open(target_file, 'a') as tarfile:
    for filePath in all_articles:
        # filter out non-article files
        if 'article' in filePath:
            #print(filePath)
            try:
                with open(filePath, 'r') as srcfile:
                    contents = srcfile.read()
                    tarfile.write(contents + '\n')
                    count += 1
                    srcfile.close()
            except:
                print('error!')
    tarfile.close()
print('total files written: ', count)

        
```
